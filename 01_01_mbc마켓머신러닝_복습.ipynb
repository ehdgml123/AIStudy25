```python
# 도미 생선의 길이 : bream_length 도미용 길이 리스트
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
# 도미 생선의 무게  : dream_weigh 도미용 무게 리스트
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
import matplotlib.pyplot as plt  # 파이썬에서 과학 계산용 그래프를 그리는 대표적인 패키지 : matplotlib

# scatter -> 차트 종류 (x 축: 길이값(bream_length) , y : 무게(bream_weight))
plt.scatter(bream_length, bream_weight)  # 차트에 값을 넣음
plt.xlabel('length')   # x축의 이름표
plt.ylabel('weight')   # y축의 이름표
plt.show()

# 2차원 그래프에 점으로 나타낸것
```

```python

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 이 2개의 리스트를 합쳐서 ai 학습용 자료로 만들어야 함

length = bream_length + smelt_length  # 도미길이 리스트 + 방어길이 리스트
weight = bream_weight + smelt_weight  # 도미무게 리스트 + 빙어무게 리스트

print(length)
print(weight)
```

```python
# zip() 나열된 리스트 각각에서 하나씩 원소를 꺼내어 반환
fish_date = [[l,w] for l,w in zip(length, weight)]  # length, weight 를 2차원 리스트로 만듬

print(fish_date)
```

```python
# 머신러닝  은 정답을 target 이라는 값으로 알려줘야 한다.
# target 0(빙어), 1(도미) 을 처리하려 함

# 생선 49개에 대한 정답 리스트를 만들어야 함
# 머신러닝에서 정답 데이터를 2진 분류로 테스트 해볼까함 (1:도미, 0 : 빙어)
fish_target = [1]*35 + [0]*14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier # 사이킷런패키지안에. 이웃패키지
                         # k-최근접 이웃 알로리즘

kn = KNeighborsClassifier()   # 임포트 된 객체를 생성하여 변수에 연결

# kn. 하면 k-최근접 알고리즘을 사용한다
```

```python
kn.fit(fish_date, fish_target)

# fish_date , fish_target 을 전달하여 도미를 찾기위한 기준을 학습(.fit) 시킨다.
```

```python
kn.score(fish_date, fish_target)  # .score() -> kn 얼마나 잘 훈련되었는지 평가

# 1.0 -> 100%(정확도)
```

```python
kn.predict([[30,600]])   # .predict[길이, 무게]  새로운 데이터의 정답을 예측
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49)  # 이웃의 개수를 49개로 설정

kn49.fit(fish_date, fish_target)
kn49.score(fish_date, fish_target)
```
